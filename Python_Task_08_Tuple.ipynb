## 12. Find the maximum and minimum values in a tuple.
Input: (10, 20, 4, 45, 99, 98)
Output:
Max: 99  
Min: 4


```python
Input=(10, 20, 4, 45, 99, 98)
min_=Input[0]
max_=Input[0]
for i in Input:
    if i>max_:
        max_=i
    elif i<min_:
        min_=i
print("Minimum-",min_)
print("Maximum-",max_)
```
