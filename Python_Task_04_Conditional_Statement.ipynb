### 4.Write a Python script that checks if a given year is a leap year. The program should consider the rules for leap years: divisible by 4, not divisible by 100 unless also divisible by 400.
### Input1:
### Enter a year: 2024
### Output1:
### 2024 is a leap year.

### Input2:
### Enter a year: 2023
### Output2:
### 2023 is not a leap year.


```python
year=int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

### 5.Create a BMI calculator that accepts a person's height (in meters) and weight (in kilograms), calculates the BMI, and categorizes it as Underweight, Normal, Overweight, or Obese using predefined thresholds.
### Input:
### Enter weight (kg): 70
### Enter height (m): 1.75
### Output:
### Your BMI is 22.9 - Normal weight.

```python
weight= float(input("Enter weight (kg):"))
height= float(input("Enter height (m):"))
bmi= weight / (height ** 2)
print(f"Your BMI is {bmi:.1f} - ", end='')

if bmi < 18.5:
    print("Underweight")
elif 18.5<= bmi < 25:
    print("Normal weight")
elif 25<= bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

```python
print("Choose conversion:")
print("1.Celsius to Fahrenheit")
print("2.Fahrenheit to Celsius")
choice = int(input("Enter choice (1 or 2): "))

if choice== 1:
    celsius= float(input("Enter temperature in Celsius: "))
    fahrenheit= (celsius * 9/5) + 32
    print(f"{celsius}°C is equal to {fahrenheit:.1f}°F.")
    
elif choice == 2:
    fahrenheit = float(input("Enter temperature in Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"{fahrenheit}°F is equal to {celsius:.1f}°C.")
    
else:
    print("Invalid choice! Please enter 1 or 2.")
```

### 6.Create a Python-based Scientific Calculator that presents a menu-driven interface to the user. The calculator should support the following operations: 
### Basic Arithmetic Operations:
### Addition (+)
### Subtraction (-)
### Multiplication (×)
### Division (/)

### Scientific Operations:
### Power (x^y)
### Square Root (√x)
### Logarithm (base 10)
### Sine (sin x in degrees)
### Cosine (cos x in degrees)
### Tangent (tan x in degrees)
### Factorial (x!)

### Requirements:

### Ask the user to select an operation from the menu.
### Accept input values as needed (one or two numbers depending on the operation).
### Perform the selected operation and display the result.
### Handle edge cases (e.g., division by zero, negative square root, factorial of negative or non-integer numbers).
### Repeat until the user chooses to exit.

### Input:
### === Scientific Calculator ===
### 1. Addition (+)
### 2. Subtraction (-)
### 3. Multiplication (×)
### 4. Division (/)
### 5. Power (x^y)
### 6. Square Root (√x)
### 7. Logarithm (log base 10)
### 8. Sine (sin x in degrees)
### 9. Cosine (cos x in degrees)
### 10. Tangent (tan x in degrees)
### 11. Factorial (x!)
### 0. Exit

### Enter choice (0–11): 3
### Enter first number: 6
### Enter second number: 4

### Output:
### Result: 6 * 4 = 24


```python
import math

while True:
    print("\n=== Scientific Calculator ===")
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (×)")
    print("4. Division (/)")
    print("5. Power (x^y)")
    print("6. Square Root (√x)")
    print("7. Logarithm (log base 10)")
    print("8. Sine (sin x in degrees)")
    print("9. Cosine (cos x in degrees)")
    print("10. Tangent (tan x in degrees)")
    print("11. Factorial (x!)")
    print("0. Exit")

    choice = int(input("\nEnter choice (0–11): "))

    if choice == 0:
        print("Exiting... Thank you for using the calculator!")
        break

    
    if choice == 1:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print("Result:", a, "+", b, "=", a + b)

    elif choice == 2:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print("Result:", a, "-", b, "=", a - b)

    elif choice == 3:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print("Result:", a, "*", b, "=", a * b)

    elif choice == 4:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        if b == 0:
            print("Error: Division by zero is not allowed!")
        else:
            print("Result:", a, "/", b, "=", a / b)

    elif choice == 5:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))
        print("Result:", a, "^", b, "=", a ** b)

    
    elif choice == 6:
        x = float(input("Enter number: "))
        if x < 0:
            print("Error: Cannot calculate square root of a negative number!")
        else:
            print("Result: √", x, "=", math.sqrt(x))

    elif choice == 7:
        x = float(input("Enter number: "))
        if x <= 0:
            print("Error: Logarithm is defined only for positive numbers!")
        else:
            print("Result: log10(", x, ") =", math.log10(x))

    elif choice == 8:
        x = float(input("Enter number (in degrees): "))
        print("Result: sin(", x, ") =", math.sin(math.radians(x)))

    elif choice == 9:
        x = float(input("Enter number (in degrees): "))
        print("Result: cos(", x, ") =", math.cos(math.radians(x)))

    elif choice == 10:
        x = float(input("Enter number (in degrees): "))
        if (x - 90) % 180 == 0:
            print("Error: tan(", x, ") is undefined!")
        else:
            print("Result: tan(", x, ") =", math.tan(math.radians(x)))

    elif choice == 11:
        x = int(input("Enter number: "))
        if x < 0:
            print("Error: Factorial is only defined for non-negative integers!")
        else:
            print("Result:", x, "!", "=", math.factorial(x))

    else:
        print("Invalid choice! Please enter a number between 0 and 11.")
```

### 8.Write a Python program that takes a user-inputted password and evaluates its strength based on the presence of uppercase letters, lowercase letters, numbers, and special characters. 

### Input1: Enter your password: StrongPass123! 
### Output1: Password Strength: Strong 
### Input2: Enter your password: weakpass 
### Output2: Password Strength: Weak

```python
password = input("Enter your password: ")
has_upper = False
has_lower = False
has_digit = False
has_special = False

for char in password:
    if char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True
    elif char.isdigit():
        has_digit = True
    else:
        has_special = True

# Checking strength
if has_upper and has_lower and has_digit and has_special and len(password) >= 8:
    print("Password Strength: Strong")
elif (has_upper or has_lower) and (has_digit or has_special) and len(password) >= 6:
    print("Password Strength: Medium")
else:
    print("Password Strength: Weak")
```

## 9. Create a Python program that simulates an inventory management system. The system should allow the user to add items, update their quantity, remove items, and display the inventory using a dictionary.

Input : 

Choose an operation:
1. Add Item
2. Update Quantity
3. Remove Item
4. Display Inventory

Enter your choice: 1
Enter item name: Laptop
Enter quantity: 5


```python
inventory = {}

while True:
    print("\nChoose an operation:")
    print("1. Add Item")
    print("2. Update Quantity")
    print("3. Remove Item")
    print("4. Display Inventory")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        item = input("Enter item name: ")
        quantity = input("Enter quantity: ")
        inventory[item] = quantity
        print("Item added.")

    elif choice == '2':
        item = input("Enter item name: ")
        if item in inventory:
            quantity = input("Enter new quantity: ")
            inventory[item] = quantity
            print("Quantity updated.")
        else:
            print("Item not found.")

    elif choice == '3':
        item = input("Enter item name: ")
        if item in inventory:
            del inventory[item]
            print("Item removed.")
        else:
            print("Item not found.")

    elif choice == '4':
        print("\nInventory:")
        for item in inventory:
            print(item + ": " + inventory[item])

    elif choice == '5':
        print("Exiting...")
        break

    else:
        print("Invalid choice.")
```
