# 1. Create a class to represent a Bank Account

Objective:

Understand constructors, instance variables, and methods.

Define a class BankAccount that takes account_holder, balance as input. Implement methods like deposit(), withdraw(), and display_balance().

Sample Input:

acc1 = BankAccount("Arvind", 1000)

acc1.deposit(500)

acc1.withdraw(200)

acc1.display_balance()

Expected Output:

Deposited: 500

Withdrew: 200

Current Balance for Arvind: 1300


```python
class BankAccount:
    def __init__(self,account_holder,balance):
        self.account_holder=account_holder
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"Deposited:{amount}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient balance!")
        else:
            self.balance-=amount
            print(f"Withdrew:{amount}")

    def display_balance(self):
        print(f"Current Balance for {self.account_holder}:{self.balance}")

acc1 = BankAccount("Arvind", 1000)
acc1.deposit(500)
acc1.withdraw(200)
acc1.display_balance()
```

# 2. Create a class Student with a class variable to count total students.

Objective:

Understand how class variables work and how to access them.

    Create a class Student that keeps track of the total number of students using a class variable. Each time a student is created, the counter should increase by 1.

Sample Input:

s1 = Student("Raj")

s2 = Student("Priya")

Student.show_total_students()

Example output- Car:

Total Students: 2


```python
class Student:
    total_students=0

    def __init__(self,name):
        self.name=name
        Student.total_students+=1  

    @classmethod
    def show_total_students(cls):
        print(f"Total Students:{cls.total_students}")

s1=Student("Raj")
s2=Student("Priya")
Student.show_total_students()
```

# 3. Create a class Laptop to represent laptop details using static methods.

Objective:

Understand static methods and their purpose.

    Create a class Laptop with instance variables brand, ram, storage. Also define a static method info() that displays a general message: "Laptops are essential for programming."

Sample Input:

A1 = Laptop("Dell", "16GB", "512GB SSD")

A1.display_specs()

Laptop.info()

Expected Output:

Laptop: Dell, RAM: 16GB, Storage: 512GB SSD

Laptops are essential for programming.


```python
class Laptop:
    def __init__(self,brand,ram,storage):
        self.brand=brand
        self.ram=ram
        self.storage=storage

    def display_specs(self):
        print(f"Laptop: {self.brand}, RAM: {self.ram}, Storage: {self.storage}")

    @staticmethod
    def info():
        print("Laptops are essential for programming")

A1 = Laptop("Dell", "16GB", "512GB SSD")
A1.display_specs()
Laptop.info()
```

# 4. Define a class Circle that calculates area and circumference

Objective:

Practice constructors, instance methods, and instance variables.

Define a class Circle that accepts radius during object creation and has methods area() and circumference().

Sample Input:

c1 = Circle(7)

print("Area:", c1.area())

print("Circumference:", c1.circumference())

Expected Output:

Area: 153.93

Circumference: 43.98


```python
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*self.radius*self.radius
    def circumference(self):
        return 2*3.14*self.radius
c1=Circle(7)
print("Area:",c1.area())
print("Circumference:",c1.circumference())
```

# 5. Create a class Employee with class method to change company name

Objective:

Learn how to define and use class methods.

    Define a class Employee with instance variables name and role. Use a class variable company = "TCS". Add a class method change_company() to update the company name for all employees.

    Sample Input:

    e1 = Employee("Ravi", "Analyst")

    e1.display()


    Employee.change_company("Infosys")

    e1.display()


    Expected Output:

    Employee: Ravi, Role: Analyst, Company: TCS

    Employee: Ravi, Role: Analyst, Company: Infosys


```python
class Employee:
    company="TCS"
    def __init__(self,name,role):
        self.name=name
        self.role=role

    def display(self):
        print(f"Employee: {self.name}, Role: {self.role}, Company: {Employee.company}")

    @classmethod
    def change_company(cls,new_company):
        cls.company=new_company

e1 = Employee("Ravi","Analyst")
e1.display()

Employee.change_company("Infosys")
e1.display()
```
