# 3.Set Operations and Methods Write a Python program that takes a set of numbers and performs the following operations:
Add a new number to the set.

Remove a number from the set.

Check if a specific number is in the set.

Find the size of the set.


```python
numbers = {10, 20, 30, 40, 50}
print("Initial Set:", numbers)

#add a number
numbers.add(60)
print("After Adding 60:", numbers)

#remove a number
numbers.remove(20)   
print("After Removing 20:", numbers)

#check if a specific number exists
check_num = 30
if check_num in numbers:
    print(f"{check_num} is present in the set.")
else:
    print(f"{check_num} is NOT present in the set.")

#find size of the set
print("Size of set:", len(numbers))
```

# 4.Create a Python list with duplicate elements and convert it into a set to remove duplicates. Then, write a program to count how many duplicates were removed.

```python
lst = [10, 20, 30, 20, 40, 10, 50, 30, 60, 10]

print("Original List:",lst)

# Convert list to set
numbers_set = set(lst)
print("After Removing Duplicates (Set):", numbers_set)

duplicates_removed = len(lst) - len(numbers_set)
print("Number of duplicates removed:", duplicates_removed)
```

# 5.Describe Set and properties of Set with examples

### -A set is an unordered, mutable collection of unique elements in Python.

### -Eg.
   ### my_set = {1, 2, 3}

### -using the set() constructor:
   ### my_set = set([1, 2, 3])

### -Properties of Sets in Python

### -Unordered.	

### -Unique Elements.	

### -Duplicate values are automatically removed.

### -Mutable - You can add or remove items.

### -Immutable Elements Only.

```python
#Creating a Set
s = {1, 2, 3, 4}
print(s)

#Adding Elements
s.add(5)
print(s)  

#adding multiple elements
s.update([6,7,8,9])
print(s)

#Removing Elements
s.remove(2)
print(s)  

#discard
s.discard(10) 

#Set Union
a = {1, 2, 3}
b = {3, 4, 5}
print(a | b)  

#Set Intersection
print(a & b)  

#Set Difference
print(a - b)  

#Set Symmetric Difference
print(a ^ b)  
```
