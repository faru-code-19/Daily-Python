## 1.Print All Even Numbers Between 1 to 20 Using a For Loop
Objective:
Print all even numbers between 1 and 20 using a for loop.

Sample Input:

Start: 1

Expected Output:

## 3.Check if a Number is Prime Using a For Loop
Objective:

Determine whether a given number is prime or not using a for loop.

Sample Input:

Number: 29

Expected Output:
29 is a Prime Number


```python
num=int(input("Enter a number:"))
if num > 1:
    for i in range(2, num):
        if num%i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num,"is a prime number")
else:
    print(num,"is not a prime number")
```

## 4.Merge Two Lists into a Dictionary Using a For Loop
Objective:

Combine two lists into a dictionary, where one list contains keys and the other contains values.

Sample Input:

Keys:["name", "age", "city"]

Values: ["Pavan", 25, "Bangalore"]

Expected Output:

{'name': 'Pavan', 'age': 25, 'city': 'Bangalore'}


```python
keys = ["name", "age", "city"]
values = ["Pavan", 25, "Bangalore"]
result = {}

for i in range(len(keys)):
    result[keys[i]] = values[i]

print(result)
```

## 5.Find the Sum of Dictionary Values Using a For Loop

Objective:

Compute the sum of all values in a dictionary using a For loop.

Sample Input:

{‘Teja’: 50000, 'Ved': 75000, 'Pavan': 60000}

Expected Output:

Sum of Dictionary Values: 185000


```python
dct={'Teja': 50000, 'Ved': 75000, 'Pavan': 60000}
sum=0
for i in dct.values():
    sum+=i
print('Sum of Dictionary Values:',sum)
```

## 6.Flatten a Nested List Using a For Loop
Objective:

Convert a nested list into a single flattened list using a for loop.

Sample Input:

[[1, 2, 3], [4, 5], [6, 7, 8, 9]]

Expected Output:

[1, 2, 3, 4, 5, 6, 7, 8, 9]


```python
lst=[[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flatten_lst=[]

for i in lst:
    #print(i)
    for j in i:
        flatten_lst.append(j)
print(flatten_lst)
```

## 10.Print Factorials of Numbers from 1 to N Using a For Loop
Objective:

Print the factorial of each number from 1 to N using a for loop.


```python
n=int(input("Enter a number:"))
for i in range(1,n + 1):
    fact= 1
    for j in range(1,i + 1):
        fact*= j
    print(f" Factorial of {i} is {fact}")
```
