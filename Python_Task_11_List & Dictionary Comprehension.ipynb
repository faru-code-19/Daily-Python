# 1-Generate a List of Squares
Objective:

Take a number n from the user.

Create a list containing squares of numbers from 1 to n.

Expected Output Example (n=5):

[1, 4, 9, 16, 25]


```python
n=int(input("enter a number:"))
l=[i**2 for i in range(1,n+1)]
print(l)
```

# 2. Create a List of Even Numbers

Objective:

Take a number n from the user.

Generate a list of all even numbers from 1 to n.

Expected Output:

[2, 4, 6, 8]


```python
n=int(input("enter numnber:"))
l=[ i for i in range(1,n) if i%2==0]
print(l)
```

# 3. Extract Vowels from a String

Objective:

Take a string input from the user.

Create a list of all vowels present in the string using list comprehension.

    Expected Input:
	Hello World

    Expected Output:

    ['e', 'o', 'o']


```python
s=input("enter a string:")
v='aeiouAEIOU'
l=[i for i in s if i.lower() in v]
print(l)
```

# 4. Filter Positive Numbers
Objective:

Given a list of integers (both positive and negative), use list comprehension to create a new list containing only positive numbers.

    Expected Input:

[-4, 3, -2, 1, 0, -7, 8]

Expected Output:

[3, 1, 8]


```python
l=[-4, 3, -2, 1, 0, -7, 8]
l1=[i for i in l if i>0]
print(l1)
```

# 5. Create a List of Words Starting with 'a'

Objective:

Take a list of words from the user.

Using list comprehension, extract all words that start with the letter 'a' or 'A'.

Expected Input:

["apple", "banana", "anchor", "dog", "ant"]

Expected Output:

['apple', 'anchor', 'ant']


```python
l=eval(input("enter a list of words:"))
l=[i for i in l if i.lower().startswith('a')]
print(l)
```

# 6. Build a List of First Letters
Objective:

Take a list of words.

    Create a new list containing only the first letter of each word.

Expected Input:

["cat", "dog", "elephant"]

Expected Output:

['c', 'd', 'e']


```python
l=["cat", "dog", "elephant"]
l1=[i[0] for i in l]
print(l1)
```

# 7. Create a Dictionary of Numbers and Their Squares
Objective:

Take a number n from the user.

Create a dictionary where keys are numbers from 1 to n and values are their squares.

Expected Output Example (n=3):

{1: 1, 2: 4, 3: 9}


```python
n=int(input("enter a number:"))
d={i:i**2 for i in range(1,n+1)}
print(d)
```

# 8. Create a Dictionary from Two Lists
Objective:

Take two lists: one of keys and one of values.

    Create a dictionary mapping them.

    Expected Input:

    ["name", "age"] and ["Kevin", 24]

    Expected Output:


    {'name': Kevin, 'age': 24}


```python
l1=["name","age"]
l2=["kevin",24]

d=dict(zip(l1,l2))
print(d)
```

# 9. Generate a List of Words with Length > 3
Objective:

Take a sentence as input.

Create a list of words that have more than 3 characters.

Expected Input:

"This is Python comprehension"

Expected Output:

    ['This', 'Python', 'comprehension']


```python
s="This is Python comprehension"
l=s.split()
l1=[i for i in l if len(i)>3]
print(l1)
```

# 10. Extract Digits from a String
Objective:

Take a string input.

Create a list containing only digits from the string.

Expected Input:

"abc123def45"

Expected Output:

['1', '2', '3', '4', '5']



```python
s=input("enter a string:")
l=[i for i in s if i.isdecimal()]
print(l)
```

# 11. Find Length of Each Word
Objective:

Take a sentence as input.

Create a dictionary where each word is a key and its length is the value.

    Expected Input:
	
"List comprehension is cool"

Expected Output:

{'List': 4, 'comprehension': 13, 'is': 2, 'cool': 4}


```python
s=input("Enter a sentence:")
d={i:len(i) for i in s.split()}
print(d)
```
