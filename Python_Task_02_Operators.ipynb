# 2. Logical Operators
### 1.What are logical operators in Python? Where do we use them?


- Logical operators are used to perform logical operations. 
- They return boolean values (True/False) as output.
- Logical operators are:
    - and - it is used when all the conditions are true.
    - or  - it is used when either one of the condition is true.
    - not - it is used to reverse the value of a conditon.
- Logical operators are used in decision-making condition.

### 2.Create a Python script that defines two boolean variables, is_sunny and is_warm, with appropriate values (True or False).
### Use logical operators (and, or, not) to create the following conditions:
- Check if it's both sunny and warm and print the result.
- Check if it's either sunny or warm and print the result.
- Check if it's not sunny and print the result.


```python
is_sunny=True
is_warm=False

print(is_sunny and is_warm)
print(is_sunny or is_warm)
print(not is_sunny)
```
