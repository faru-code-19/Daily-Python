```python
class BankAccount:
    bank_name="Faru_Bank"

    def __init__(self,acc_no,name,pin,balance=0):
        self.acc_no=acc_no
        self.name=name
        self.__pin=pin        # private variable
        self.balance=balance

    def check_pin(self):
        return int(input("Enter PIN:"))==self.__pin

    def show_details(self):
        if self.check_pin():
            print("\nBank:",BankAccount.bank_name)
            print("Name:",self.name)
            print("Account No:",self.acc_no)
            print("Balance: ₹",self.balance)
        else:
            print("Wrong PIN")

    def deposit(self):
        if self.check_pin():
            amount=float(input("Enter amount: "))
            self.balance+=amount
            print("Deposit successful. Balance:",self.balance)

    def withdraw(self):
        if self.check_pin():
            amount=float(input("Enter amount: "))
            if amount<=self.balance:
                self.balance-=amount
                print("Withdrawal successful. Balance:",self.balance)
            else:
                print("Insufficient balance")

    @staticmethod
    def thank_you():
        print("\nThank you for choosing Faru_Bank")


# --------- MENU DRIVEN----------
accounts={}

while True:
    print("\n===== BANK MENU =====")
    print("1.Create Account")
    print("2.Show Details")
    print("3.Deposit")
    print("4.Withdraw")
    print("5.Exit")

    choice=input("Enter choice: ")

    if choice=="1":
        acc_no=input("Account Number:")
        name=input("Name:")
        pin=int(input("PIN:"))
        balance=float(input("Balance:"))
        accounts[acc_no]=BankAccount(acc_no, name, pin, balance)
        print("Account created")

    elif choice in ["2","3","4"]:
        acc_no=input("Enter Account Number:")

        if acc_no not in accounts:
            print("Account not found")
            continue

        acc=accounts[acc_no]

        if choice == "2":
            acc.show_details()
        elif choice == "3":
            acc.deposit()
        elif choice == "4":
            acc.withdraw()

    elif choice == "5":
        BankAccount.thank_you()
        break

    else:
        print("Invalid choice")
```
