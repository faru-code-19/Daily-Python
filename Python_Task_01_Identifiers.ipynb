# 1.Explain about identifiers and their rules.

Identifiers are the names given to a variable, function, classes, objects, module.

#### RULES FOR IDENTIFIERS

1- It can contain Alphabets (A-Z), (a-z), Numbers(0-9) and underscore(_).

2- It can start with Alphabets (A-Z), (a-z) or with an underscore(_) only.

3- It cannot start with a Number or a Digit (0-9).

4- Other special characters such as (!,@,#,$,^,&,*,(,),-,=,+) cannot be used, only underscore(_) can be used.

5- We cannot use Reserved Keywords as Identifiers.

6- Case sensitive- Python and python both are different.

# 3.Explain why Python treats some identifiers as invalid.

#### Python follows strict rules for naming identifiers (variables, functions, classes, etc.). If an identifier breaks those rules, Python treats some

#### identifiers as invalid when they:

- Start with a digit.

- Contain spaces.

- Use special characters except underscore ( _ ). 

- Match a reserved keyword.
