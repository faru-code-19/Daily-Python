# 1. Create a class for Arithmetic Operations

Objective:
Practice defining instance methods for basic operations.

Task:

Create a class Arithmetic that includes methods for performing addition, subtraction, multiplication, and division.	

Sample Input:

calc = Arithmetic()

print("Add:", calc.addition(10, 5))
                                            
print("Subtract:", calc.subtraction(10, 5))

print("Multiply:", calc.multiplication(10, 5))

print("Divide:", calc.division(10, 5))

Expected Output:

Add: 15
                                            
Subtract: 5

Multiply: 50
                                            
Divide: 2.0


```python
class Arithmetic:
    def addition(self,a,b):
        return a+b

    def subtraction(self,a,b):
        return a-b

    def multiplication(self,a,b):
        return a*b

    def division(self,a,b):
        if b==0:
            return"Error: Division by zero"
        return a/b

calc=Arithmetic()
print("Add:",calc.addition(10,5))
print("Subtract:",calc.subtraction(10,5))
print("Multiply:",calc.multiplication(10,5))
print("Divide:",calc.division(10,5))
```

# 2. Implement Single Inheritance with Vehicle and Car

Objective:

Understand how to inherit from a base class and extend it.

Task:

Define a base class Vehicle with attributes color and model. Create a derived class Car that adds an attribute num_doors. Include a __str__() method to display car details.

Sample Input:

c1 = Car("Red", "MG Hector", 4)

print(c1)


Example output:

Car: MG Hector, Color: Red, Doors: 4


```python
class Vehicle:
    def __init__(self,color,model):
        self.color=color
        self.model=model

class Car(Vehicle):
    def __init__(self,color,model,num_doors):
        super().__init__(color,model)  
        self.num_doors=num_doors

    def __str__(self):
        return f"Car: {self.model}, Color: {self.color},Doors: {self.num_doors}"

c1=Car("Red","MG Hector",4)
print(c1)
```

# 3. Implement Multiple Inheritance with Person and Address

Objective:

Practice using multiple inheritance to combine data from two different base classes.


Task:

Define two base classes:

Person with attributes name and age

Address with attributes city and country

Create a derived class Employee that inherits from both and adds an attribute job. Include a __str__() method to show full employee details.


Sample Input:

emp1 = Employee("Sirisha", 25, "Delhi", "India", "Data Scientist")

print(emp1)

Expected Output:

Employee: Sirisha, Age: 25, City: Delhi, Country: India, Job: Data Scientist


```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

class Address:
    def __init__(self,city,country):
        self.city=city
        self.country=country

class Employee(Person,Address):
    def __init__(self,name,age,city,country,job):
        Person.__init__(self,name,age)
        Address.__init__(self,city,country)
        self.job=job

    def __str__(self):
        return (f"Employee: {self.name}, Age: {self.age}, City: {self.city}, Country: {self.country}, Job: {self.job}")
        
emp1 = Employee("Sirisha",25,"Delhi","India","Data Scientist")
print(emp1)
```
