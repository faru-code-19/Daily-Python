# 1. Encapsulation – FlightBooking

Objective:

Understand how to use private variables to encapsulate sensitive booking information.

    Scenario:

Create a class FlightBooking with private attributes __passenger_name, __flight_number, and __amount_paid. Include methods to display booking details and apply a promo code discount.

Sample Input:

booking1 = FlightBooking("Amit", "AI302", 8000)

booking1.apply_promo("NEW1000")

booking1.show_booking()

Expected Output:

Promo code applied: ₹1000 off

Passenger: Amit | Flight: AI302 | Amount Paid: ₹7000


```python
class FlightBooking:
    def __init__(self,name,flt_number,amount):
        self.__passenger_name=name
        self.__flight_number=flt_number
        self.__amount_paid=amount

    def apply_promo(self,code):
        discount=1000
        self.__amount_paid-=discount
        print(f"Promo code applied: ₹1000 off")
        
    def show_booking(self):
        print(f"Passenger: {self.__passenger_name} | Flight: {self.__flight_number} | Amount Paid: {self.__amount_paid}")

booking1 = FlightBooking("Amit", "AI302", 8000)
booking1.apply_promo("NEW1000")
booking1.show_booking()
```

# 2. Class Variable – Library

Objective:

Demonstrate the use of a class variable to track the total number of books added to the library.

Scenario:

Create a class Library with instance attributes like title and author, and a class variable total_books. Increment total_books every time a book is added. Add a class method to display the total number of books.

Sample Input:

b1 = Library("Atomic Habits", "James Clear")

b2 = Library("Deep Work", "Cal Newport")

Library.display_total_books()


Example output:

Total books in the library: 2


```python
class Library:
    total_books=0

    def __init__(self,title,author):
        self.title=title
        self.author=author
        Library.total_books+=1

    @classmethod
    def display_total_books(cls):
        print(f"Total books in the library: {cls.total_books}")

b1 = Library("Atomic Habits", "James Clear")
b2 = Library("Deep Work", "Cal Newport")
Library.display_total_books()
```

# 3. Static Method – Smartphone

Objective:

Understand the purpose and usage of static methods in a real-world context.

Scenario:

Create a class Smartphone with attributes brand, model, and price. Include a static method general_info() that returns general information about smartphones.

Sample Input:

s1 = Smartphone("Samsung", "Galaxy S21", 50000)

s1.show_details()

Smartphone.general_info()

Expected Output:

Brand: Samsung | Model: Galaxy S21 | Price: ₹50000

Smartphones are portable devices used for communication, entertainment, and work.


```python
class Smartphone:
    def __init__(self,brand,model,price):
        self.brand=brand
        self.model=model
        self.price=price
    

    def show_details(self):
        print(f"Brand: {self.brand} | Model: {self.model} | Price: {self.price}")
        
    @staticmethod    
    def general_info():
        print(f"Smartphones are portable devices used for communication, entertainment, and work.")

s1 = Smartphone("Samsung", "Galaxy S21", 50000)
s1.show_details()
Smartphone.general_info()
```

# 4. Inheritance – Vehicle and ElectricCar

Objective:

Practice basic inheritance and method overriding using a real-world vehicle system.

    Scenario:

Create a base class Vehicle with brand, model attributes and a method start_engine() that prints a general start message.

    Then create a subclass ElectricCar that overrides start_engine() to print a message specific to electric vehicles.


Sample Input:

v1 = Vehicle("Toyota", "Fortuner")

v1.start_engine()

e1 = ElectricCar("Tesla", "Model S")

e1.start_engine()

Expected Output:

Starting Toyota Fortuner engine...

Powering up Tesla Model S electric system...


```python
class Vehicle:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
        
    def start_engine(self):
        print(f"Starting {self.brand} {self.model} engine...")

class ElectricCar(Vehicle):
    
    def start_engine(self):
        print(f"Powering up {self.brand} {self.model} electric system...")

v1 = Vehicle("Toyota", "Fortuner")
v1.start_engine()
e1 = ElectricCar("Tesla", "Model S")
e1.start_engine()
```

# 5. Inheritance – Appliance and WashingMachine

Objective:

Learn how a subclass can inherit common behavior and add specific functionality.

    Scenario:

Create a class Appliance with a method turn_on().

Create a subclass WashingMachine that adds a method start_wash_cycle() and uses turn_on() from the parent.

Sample Input:

a1 = WashingMachine("LG", "TurboWash")

a1.turn_on()

a1.start_wash_cycle()

Expected Output:

LG TurboWash is now turned on.

Starting wash cycle in LG TurboWash...


```python
class Appliance:    
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
    
    def turn_on(self):
        print(f"{self.brand} {self.model} is now turned on")
        
class WashingMachine(Appliance):
    def start_wash_cycle(self):
        print(f"Starting wash cycle in {self.brand} {self.model}...")

a1 = WashingMachine("LG", "TurboWash")
a1.turn_on()           
a1.start_wash_cycle()  
```
